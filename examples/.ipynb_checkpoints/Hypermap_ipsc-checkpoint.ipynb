```python
adata=sc.read_h5ad('../ipsc_hvg_2500_norm.h5ad')
adata.X=adata.X.toarray()
adata.obs['context_cell']=adata.obs['Cell_Line'].str.split('_',expand=True)[0]
pert_test = adata.obs['condition'][adata.obs['condition']!='ctrl'].unique()
```

```python
donor_means = {} 
ctrl_mask = adata.obs['condition'] == 'ctrl'

for donor in adata.obs['context_cell'].unique():
    donor_ctrl_mask = (adata.obs['context_cell'] == donor) & ctrl_mask
    donor_means[donor] = np.mean(adata[donor_ctrl_mask].X, axis=0)
 
print("Computing normalized expressions...")

normalized_exp = np.zeros_like(adata.X)

for donor in adata.obs['context_cell'].unique():
    donor_mask = adata.obs['context_cell'] == donor
    normalized_exp[donor_mask] = adata[donor_mask].X - donor_means[donor] ## All donor cells- mean

adata.layers['delta'] = normalized_exp
```

### Fig 2a

```python
# ── 1. Perturbation response similarity (mean cell-to-cell correlation) ────────

cells = adata.obs['context_cell'].unique()
conditions = [c for c in adata.obs['condition'].unique() if c != 'ctrl']

condition_corrs = {}
for condition in conditions:
    condition_cells = {
        cell: (adata.obs['context_cell'] == cell) & (adata.obs['condition'] == condition)
        for cell in cells
    }
    condition_cells = {k: v for k, v in condition_cells.items() if np.sum(v) > 0}
    if len(condition_cells) < 2:
        continue

    cell_list = list(condition_cells.keys())
    delta_matrix = np.column_stack([
        np.mean(adata[idx].layers['delta'], axis=0)
        for idx in condition_cells.values()
    ])
    corr = np.corrcoef(delta_matrix.T)
    condition_corrs[condition] = pd.DataFrame(corr, index=cell_list, columns=cell_list)

mean_corr = pd.DataFrame(0.0, index=cells, columns=cells)
count_matrix = pd.DataFrame(0, index=cells, columns=cells)
for corr_df in condition_corrs.values():
    mean_corr.loc[corr_df.index, corr_df.columns] += corr_df
    count_matrix.loc[corr_df.index, corr_df.columns] += 1
mean_corr = mean_corr / count_matrix.where(count_matrix > 0)

# ── 2. Control state similarity (e-distance → RBF kernel) ─────────────────────


ac = adata[adata.obs['condition'] == 'ctrl'].copy()
adat_sub = scperturb.equal_subsampling(ac, 'context_cell', N_min=500)
sc.pp.pca(adat_sub, use_highly_variable=True)
sc.pp.neighbors(adat_sub)
estats = scperturb.edist(adat_sub, obs_key='context_cell', obsm_key='X_pca',
                         dist='sqeuclidean')

sigma = estats.values.std()
similarity_df = np.exp(-estats**2 / (2 * sigma**2))
similarity_df.index.name = None

# ── 3. Combine into long-form table ───────────────────────────────────────────

mean_corr_01 = (mean_corr + 1) / 2

out = (mean_corr_01
       .reset_index().rename(columns={'index': 'd1'})
       .melt(id_vars='d1', var_name='d2', value_name='pert_response'))

out2 = (similarity_df
        .reset_index().rename(columns={'index': 'd1'})
        .melt(id_vars='d1', var_name='d2', value_name='ctrl_similairty'))

combined_scores = (out2
                   .merge(out, on=['d1', 'd2'])
                   .query('d1 != d2'))

# ── 4. Assign quadrants ───────────────────────────────────────────────────────

x_thr, y_thr = 0.5, 0.5

def assign_quad(row):
    xc, yc = row['ctrl_similairty'], row['pert_response']
    if   xc >= x_thr and yc >= y_thr: return 'Q1: High ctrl • High resp'
    elif xc <  x_thr and yc >= y_thr: return 'Q2: Low ctrl • High resp'
    elif xc <  x_thr and yc <  y_thr: return 'Q3: Low ctrl • Low resp'
    else:                              return 'Q4: High ctrl • Low resp'

combined_scores['quadrant'] = combined_scores.apply(assign_quad, axis=1)

# ── 5. Plot ───────────────────────────────────────────────────────────────────

x = combined_scores['ctrl_similairty']
y = combined_scores['pert_response']

fig, ax = plt.subplots(figsize=(7, 7))

ax.scatter(x, y, color='#ce4257', s=80, alpha=0.4,
           edgecolor='white', linewidth=0.3)

ax.axvline(x_thr, color='grey', linestyle='--', linewidth=1)
ax.axhline(y_thr, color='grey', linestyle='--', linewidth=1)

quad_labels = {
    'Q1: High ctrl • High resp': (x_thr + 0.02, y_thr + 0.02),
    'Q2: Low ctrl • High resp':  (x.min() + 0.02, y_thr + 0.02),
    'Q3: Low ctrl • Low resp':   (x.min() + 0.02, y.min() + 0.02),
    'Q4: High ctrl • Low resp':  (x_thr + 0.02, y.min() + 0.02),
}
for label, (lx, ly) in quad_labels.items():
    ax.text(lx, ly, label, fontsize=10)

ax.set_xlim(-0.01, 1)
ax.set_ylim(-0.01, 1)
ax.set_xlabel('Control state similarity', fontsize=12)
ax.set_ylabel('Perturbation response similarity', fontsize=12)

plt.tight_layout()
plt.savefig('Figures/Fig1a.svg')
plt.show()
```

### Fig 2b

```python
def _compute_ctrl_pca_and_project_delta(adata_pair, pert_name,
                                         donor_key, condition_key,
                                         ctrl_label, delta_layer):
    """Returns (donors, donor_colors, pc_ctrl, ad_ctrl, pc_delta, ad_pert)."""
    donors = adata_pair.obs[donor_key].unique().tolist()
    assert len(donors) == 2
    d1, d2 = donors
    donor_colors = {d1: "#474973", d2: "#f27059"}

    # --- Control PCA ---
    ad_ctrl = adata_pair[adata_pair.obs[condition_key] == ctrl_label].copy()
    sc.pp.scale(ad_ctrl)
    sc.tl.pca(ad_ctrl, n_comps=2)
    pcs = ad_ctrl.varm["PCs"][:, :2]
    pc_ctrl = ad_ctrl.obsm["X_pca"][:, :2]

    # Remove outliers
    z1 = (pc_ctrl[:, 0] - pc_ctrl[:, 0].mean()) / pc_ctrl[:, 0].std()
    z2 = (pc_ctrl[:, 1] - pc_ctrl[:, 1].mean()) / pc_ctrl[:, 1].std()
    keep = (np.abs(z1) < 3) & (np.abs(z2) < 3)
    pc_ctrl = pc_ctrl[keep]
    ad_ctrl = ad_ctrl[keep].copy()

    # --- Delta projection ---
    ad_pert = adata_pair[adata_pair.obs[condition_key] == pert_name].copy()
    delta = ad_pert.layers[delta_layer]
    delta_z = (delta - delta.mean(0)) / (delta.std(0) + 1e-9)
    pc_delta = (delta_z @ pcs)[:, :2]

    return donors, donor_colors, pc_ctrl, ad_ctrl, pc_delta, ad_pert


def plot_figure2b(
    adata,
    pairs,                        # list of (donor_list, pert_name) tuples
    donor_key="context_cell",
    condition_key="condition",
    ctrl_label="ctrl",
    delta_layer="delta",
    plot_contours=True,
    save_path=None
):
    """
    2×2 figure (Figure 2b).
    Each row = one donor pair; left col = ctrl PCA, right col = Δ-PCA.
    """
    sns.set_style("white")
    n_rows = len(pairs)
    fig, axes = plt.subplots(n_rows, 2, figsize=(12, 6 * n_rows))

    for row, (donor_list, pert_name) in enumerate(pairs):
        adata_pair = adata[adata.obs[donor_key].isin(donor_list)].copy()

        donors, donor_colors, pc_ctrl, ad_ctrl, pc_delta, ad_pert = \
            _compute_ctrl_pca_and_project_delta(
                adata_pair, pert_name, donor_key, condition_key,
                ctrl_label, delta_layer
            )

        # Shared axis limits
        global_xmin = min(pc_ctrl[:, 0].min(), pc_delta[:, 0].min()) - 3
        global_xmax = max(pc_ctrl[:, 0].max(), pc_delta[:, 0].max()) + 3
        global_ymin = min(pc_ctrl[:, 1].min(), pc_delta[:, 1].min()) - 3
        global_ymax = max(pc_ctrl[:, 1].max(), pc_delta[:, 1].max()) + 3

        # Left: Control PCA
        ax_ctrl = axes[row, 0]
        for d in donors:
            mask = ad_ctrl.obs[donor_key] == d
            sns.kdeplot(x=pc_ctrl[mask, 0], y=pc_ctrl[mask, 1],
                        levels=5 if plot_contours else 1,
                        fill=False, linewidths=1.2,
                        color=donor_colors[d], label=f"{d} ctrl", ax=ax_ctrl)
        ax_ctrl.set(title="Control PCA", xlabel="PC1", ylabel="PC2",
                    xlim=(global_xmin, global_xmax), ylim=(global_ymin, global_ymax))
        ax_ctrl.legend(frameon=False)

        # Right: Delta PCA
        ax_pert = axes[row, 1]
        for d in donors:
            mask = ad_pert.obs[donor_key] == d
            sns.kdeplot(x=pc_delta[mask, 0], y=pc_delta[mask, 1],
                        levels=5 if plot_contours else 1,
                        fill=False, linewidths=1.2,
                        color=donor_colors[d], label=f"{d} Δ", ax=ax_pert)
        ax_pert.set(title=f"Δ-PCA (z-score Δ) — {pert_name}",
                    xlabel="PC1", ylabel="PC2",
                    xlim=(global_xmin, global_xmax), ylim=(global_ymin, global_ymax))
        ax_pert.legend(frameon=False)

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, bbox_inches="tight")
    plt.show()

# ── Figure 2b ─────────────────────────────────────────────────────────────────

plot_figure2b(
    adata,
    pairs=[
        (['paab', 'jejf'], "INTS1"),
        (['eipl', 'kolf'], "INTS1"),
    ],
    save_path='Figures/Fig2b.svg'
)
```

### Fig 2j

```python
top_20_df.head()
```

```python
low_corr_five=top_20_df[['Donor','Value']].groupby('Donor').mean().sort_values('Value').index[:5]
```

```python
df_corr = (top_20_df[top_20_df['Donor'].isin(low_corr_five)]
           [['Gene', 'Value']]
           .groupby('Gene')
           .median())

# ── 3. High / low gene splits ─────────────────────────────────────────────────

q25 = df_corr['Value'].quantile(0.25)
q75 = df_corr['Value'].quantile(0.75)
low_genes  = df_corr[df_corr['Value'] <= q25].index.tolist()
high_genes = df_corr[df_corr['Value'] >= q75].index.tolist()
print(f"Low group: {len(low_genes)} genes | High group: {len(high_genes)} genes")

# ── 4. GO enrichment ──────────────────────────────────────────────────────────

enr_low  = gp.enrichr(gene_list=low_genes,  gene_sets=['GO_Biological_Process_2023'],
                       organism='Human', cutoff=0.05)
enr_high = gp.enrichr(gene_list=high_genes, gene_sets=['GO_Biological_Process_2023'],
                       organism='Human', cutoff=0.05)

df_low  = enr_low.results[["Term", "Adjusted P-value", "Overlap"]].copy()
df_high = enr_high.results[["Term", "Adjusted P-value", "Overlap"]].copy()
df_low["score_low"]   = -np.log10(df_low["Adjusted P-value"].replace(0, 1e-300))
df_high["score_high"] = -np.log10(df_high["Adjusted P-value"].replace(0, 1e-300))

merged = pd.merge(df_high, df_low, on="Term", how="outer", suffixes=("_high", "_low")).fillna(0)

# ── 5. Parse overlap & filter significant terms ───────────────────────────────

def parse_overlap(s):
    try:
        a, b = str(s).split("/")
        return int(a), int(b)
    except Exception:
        return 0, 0

merged[["overlap_k_low",  "term_K_low"]]  = merged["Overlap_low"].apply(parse_overlap).apply(pd.Series)
merged[["overlap_k_high", "term_K_high"]] = merged["Overlap_high"].apply(parse_overlap).apply(pd.Series)
merged["overlap_k"] = merged[["overlap_k_low", "overlap_k_high"]].max(axis=1).astype(int)
merged["term_K"]    = merged[["term_K_low",  "term_K_high"]].max(axis=1).astype(int)

thr_p = 1e-3
thr   = -np.log10(thr_p)
sig   = merged[(merged["score_high"] >= thr) | (merged["score_low"] >= thr)].copy()
if sig.empty:
    raise SystemExit("No significant GO terms at the chosen threshold.")

sig["score_max"] = sig[["score_high", "score_low"]].max(axis=1)
sig = sig.sort_values("score_max", ascending=False).head(12).copy()

# ── 6. Term label helpers ─────────────────────────────────────────────────────

def simplify_go(term):
    t = re.sub(r"\s*\(GO:\d+\)\s*$", "", str(term))
    t = (t.replace("RNA Polymerase II", "Pol II")
          .replace("Deoxyribonucleic Acid-templated", "DNA-templated")
          .replace("DNA-templated", "DNA")
          .replace("Positive Regulation Of", "↑")
          .replace("Negative Regulation Of", "↓")
          .replace("Regulation Of", "Reg.")
          .replace("Via Spliceosome", "via spliceosome")
          .replace("Procorrdfcessing", "proc.")
          .replace("Metabolic Process", "metab. proc.")
          .replace("Transcription", "Tx")
          .replace("Translation", "Tl"))
    t = t.replace(" Of ", " ").replace(" By ", " ").replace("  ", " ").strip()
    t = t.replace("Reg. Tx Initiation Pol II", "Pol II initiation (reg.)")
    t = t.replace("Reg. Tx Elongation Pol II", "Pol II elongation (reg.)")
    t = t.replace("Gene Expression", "Gene expression")
    return t

label_map = {
    "RNA proc.":                         "RNA processing",
    "Pol II elongation (reg.)":          "Pol II elongation",
    "snRNA proc.":                       "snRNA processing",
    "RNA 3'-End proc.":                  "RNA 3' processing",
    "Macromolecule Biosynthetic Process":"Macromolecule",
    "Tl":                                "Translation",
    "Gene expression":                   "Gene expression",
    "mRNA Splicing, via spliceosome":    "mRNA splicing",
    "mRNA proc.":                        "mRNA processing",
    "↑ Tx Initiation Pol II":            "Tx initiation",
    "↑ DNA Tx Initiation":               "DNA Tx initiation",
    "Pol II initiation (reg.)":          "Pol II initiation",
    "RNA Splicing, Via Transesterification Reactions With Bulged Adenosine As Nucleophile":'RNA Splicing'
}

sig["Term_simple"] = sig["Term"].apply(simplify_go)
sig["Term_label"]  = sig.apply(
    lambda r: f"{label_map.get(r['Term_simple'], r['Term_simple'])} ({r['overlap_k']}/{r['term_K']})"
    if r["term_K"] > 0 else label_map.get(r["Term_simple"], r["Term_simple"]),
    axis=1
)

# ── 7. Categorical sort (high-only → both → low-only) ────────────────────────

sig["p_low"]  = 10 ** (-sig["score_low"].astype(float))
sig["p_high"] = 10 ** (-sig["score_high"].astype(float))
sig["low_enr"]  = sig["p_low"]  <= thr_p
sig["high_enr"] = sig["p_high"] <= thr_p

def _cat(row):
    if row["low_enr"] and not row["high_enr"]: return 2
    if row["low_enr"] and row["high_enr"]:     return 1
    if not row["low_enr"] and row["high_enr"]: return 0
    return 3

sig["cat"]    = sig.apply(_cat, axis=1)
sig["p_best"] = sig[["p_low", "p_high"]].min(axis=1)
sig = sig.sort_values(["cat", "p_best"], ascending=[True, True]).reset_index(drop=True)

# ── 8. Plot ───────────────────────────────────────────────────────────────────

ypos  = np.arange(len(sig))
sizes = 40 + 25 * np.sqrt(sig["overlap_k"].clip(lower=1))

fig, ax = plt.subplots(figsize=(5.0, 6))

ax.scatter(sig["p_low"],  ypos, s=sizes, c="white", edgecolor="black", linewidths=0.8, zorder=3)
ax.scatter(sig["p_high"], ypos, s=sizes, c="black", edgecolor="black", linewidths=0.8, zorder=3)

ax.axvline(0.001, ls="--", lw=0.9, color="#7a7a7a", zorder=2)
ax.set_yticks(ypos)
ax.set_yticklabels(sig["Term_label"], fontsize=8)
ax.set_xscale("log")
ax.xaxis.set_major_formatter(mticker.FuncFormatter(
    lambda x, _: r"$10^{0}$" if x == 1
    else rf"$10^{{{int(round(math.log10(x)))}}}$"
))
ax.set_xlabel("adj p-value", fontsize=10)
ax.set_title("GO BP enrichment: High vs Low", fontsize=11)

pmin = float(np.nanmin(sig[["p_low", "p_high"]].to_numpy()))
ax.set_xlim(9, pmin * 0.35)
ax.grid(axis="y", color="#e5e5e5", lw=0.6)
ax.grid(False, axis="x")

plt.tight_layout()
plt.savefig("Figures/Fig2j.svg")
plt.show()
```
